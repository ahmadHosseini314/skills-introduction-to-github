```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def double_pendulum(y, t, L1, L2, m1, m2, g):
    theta1, omega1, theta2, omega2 = y

    # Equations of motion
    dydt = [omega1,
            (-g * (2 * m1 + m2) * np.sin(theta1) - m2 * g * np.sin(theta1 - 2 * theta2) -
             2 * np.sin(theta1 - theta2) * m2 * (omega2 ** 2 * L2 + omega1 ** 2 * L1 * np.cos(theta1 - theta2))) /
            (L1 * (2 * m1 + m2 - m2 * np.cos(2 * theta1 - 2 * theta2))),
            omega2,
            (2 * np.sin(theta1 - theta2) * (omega1 ** 2 * L1 * (m1 + m2) + g * (m1 + m2) * np.cos(theta1) +
             omega2 ** 2 * L2 * m2 * np.cos(theta1 - theta2))) /
            (L2 * (2 * m1 + m2 - m2 * np.cos(2 * theta1 - 2 * theta2)))]

    return dydt

# Parameters
L1 = 1.0  # Length of pendulum 1
L2 = 1.0  # Length of pendulum 2
m1 = 1.0  # Mass of pendulum 1
m2 = 1.0  # Mass of pendulum 2
g = 9.81  # Acceleration due to gravity

# Initial conditions: theta1, omega1, theta2, omega2
y0 = [np.pi / 2, 0, np.pi / 2, 0]

# Time points
t = np.linspace(0, 20, 1000)

# Solve the ODE
sol = odeint(double_pendulum, y0, t, args=(L1, L2, m1, m2, g))

# Extracting results
theta1 = sol[:, 0]
theta2 = sol[:, 2]

# Plotting the motion
plt.figure(figsize=(8, 6))
plt.plot(L1 * np.sin(theta1), -L1 * np.cos(theta1), label='Pendulum 1')
plt.plot(L2 * np.sin(theta2) + L1 * np.sin(theta1), -L2 * np.cos(theta2) - L1 * np.cos(theta1), label='Pendulum 2')
plt.title('Double Pendulum Motion')
plt.xlabel('x-position')
plt.ylabel('y-position')
plt.legend()
plt.grid()
plt.show()
```
